```python
for df_1 in [county_2015, cf]:
    df_1.set_index(["County", "State"], inplace=True)
    
cf = cf[cf['cand']=="Donald Trump"]
cf = cf[['pct']]
cf.dropna(inplace=True)
cf.head(4)
```
